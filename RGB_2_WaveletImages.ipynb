```python
import os
import cv2
import numpy as np
import pywt
from pathlib import Path

def wavelet_transform(image_path):
    """
    Apply wavelet transform to an image.

    Args:
        image_path (str): Path to the input image.

    Returns:
        np.ndarray: Wavelet transformed image.
    """
    # Read the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Cannot read image: {image_path}")
    
    # Perform wavelet decomposition using Haar wavelet
    coeffs2 = pywt.dwt2(img, 'haar')
    cA, (cH, cV, cD) = coeffs2
    
    # Combine coefficients into a single visual representation
    transformed_img = np.vstack([
        np.hstack([cA, cH]),
        np.hstack([cV, cD])
    ])
    return transformed_img

def process_and_stack_images(input_dir, output_dir):
    """
    Process images by applying wavelet transform and stacking the original and transformed images vertically.

    Args:
        input_dir (str): Path to the input dataset directory.
        output_dir (str): Path to the output directory to save stacked images.
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    
    for img_path in input_dir.rglob("*.*"):  # Iterate through all files
        if img_path.suffix.lower() in [".jpg", ".jpeg", ".png", ".bmp", ".tif"]:
            # Maintain parent folder structure
            relative_path = img_path.relative_to(input_dir)
            output_path = output_dir / relative_path
            output_path.parent.mkdir(parents=True, exist_ok=True)
            
            try:
                # Read original image
                original_img = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)
                
                # Apply wavelet transformation
                transformed_img = wavelet_transform(str(img_path))
                
                # Normalize transformed image for display
                normalized_transformed_img = cv2.normalize(transformed_img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
                
                # Stack the original and transformed images vertically
                stacked_img = np.vstack([original_img, normalized_transformed_img])
                
                # Save the stacked image
                cv2.imwrite(str(output_path), stacked_img)
                print(f"Processed and saved: {output_path}")
            except Exception as e:
                print(f"Failed to process {img_path}: {e}")

# Example usage
input_dataset_path =r"C:\Users\Subham Gupta\Desktop\Work\i2v_system_Visiontask\vehicle_dataset\train"
output_dataset_path =r"C:\Users\Subham Gupta\Desktop\Work\i2v_system_Visiontask\vehicle_dataset\train_wavelet"
process_and_stack_images(input_dataset_path, output_dataset_path)
# Ensure the input and output paths are set correctly before running the script.
# Note: Replace "path/to/input/images" and "path/to/output/images" with actual paths.
 
```
